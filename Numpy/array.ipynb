Accessing Array elements, rows, columns etc

```python
a = np.array([[12,23,34,5,3,54,2], [2,3,4,34,5,64,23]])
print(a)
a.size
a.shape
```

```python
#Get a specific element
a[0,3]
a[0,-2] #Prints the 2nd last element of 0th row
```

```python
#Get a specific row
a[0, :]
```

```python
#Get a specific column
a[:,0]
```

```python
#Getting [startIndex:EndIndex:stepSize]
a[0, 1:6:2]
```

```python
#Updating single element
a[0,3] = 25
print(a)
```

```python
#Updating all column elements with same values
a[:, 2] = 5 # Updates the elements of 3rd column
print(a)
```

```python
#Updating all column elements with different values
a[:, 3] = [7, 8]
print(a)
```

```python
#All 0s matrix
d = np.zeros(5)
d = np.zeros((5,2)) #5 rows, 2 columns
print(d)
```

```python
#All 1s matrix
d = np.ones((3,2))
print(d)
```

```python
#Initialize matrix with any number
d = np.full((2,2), 5)
print(d)
```

```python
#Initialize with any other number -> use shape of another array
d = np.full_like(a, 4)
print(d)
```

```python
#Random decimal number
d = np.random.rand(4,3)
print(d)
```

```python
#Random decimal number using shape of another array
d = np.random.random_sample(a.shape)
print(d)
```

```python
#Random integer values
d = np.random.randint(10, size=(3,3))  #10 is max int value
d = np.random.randint(5, 10, size=(3,3)) # range is 5 to 10
print(d)
```

```python
#Identity matrix
d= np. identity(5)
print(d)
```

```python
#Repeat
d = np.array([[1,2,3]])
r1 = np.repeat(d, 4, axis = 0) #d is repeated 4 times
print(r1)
```

```python
#Building a matrix
output = np.ones((5,5))
print(output)
innerMat = np.zeros((3,3))
print(innerMat)
innerMat[1,1] = 9
print(innerMat)
#Merge output and innerMat
output[1:4, 1:4] = innerMat
print(output)
```

```python
#array copy
# a = d
# a[0,1] = 5
# print(d) # d should not be changed, but d is changing here
#So we should use .copy() function
a = d.copy()
a[0,1] = 15
print(d) # d is not changing
print(a)
```
