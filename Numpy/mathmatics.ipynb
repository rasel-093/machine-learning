Linear Algebra

```python
#Matrix multiplication
mat1 = np.full((3,3), 2)
mat2 = np.full((3,3), 3)
np.matmul(mat1,mat2)
```

```python
#Determinant
arr = np.array([[2,3],[5,4]]) 
# arr.shape
det = np.linalg.det(arr)
print(det)
```

```python
#Get maximum value
np.max(stats)
```

Reorganizing Arrays

```python
before = np.array([[1,2,3,4],[6,7,8,9]])
print(before)
after = before.reshape((8,1)) # 8 rows, 1 colums
print(after)
```

```python
#Vertically stacking vectors
v1 = np.array([1,2,3,4])
v2 = np.array([6,7,8,9])
combine = np.vstack([v2,v1, v1,v2])  #Combined to one array of two rows
print(combine)
```

```python
#Horizontal stack
combine = np.hstack((v1,v2, v1, v2))
print(combine) #Combined to single column
```

```python
#Vector dot product
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
dot = np.dot(arr1,arr2)
print(dot)
```

```python
#dot product using @
dot = arr1 @ arr2
print(dot)
```
